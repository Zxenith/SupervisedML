```python
#COST FUNCTION
def cost_func(x, y, w, b):
    m = x.shape[0]
    
    cost = 0
    
    for i in range(m):
        f_wb = w * x[i] + b
        cost += (f_wb - y[i]) ** 2
    cost /= (2*m)
    
    return cost
```

```python
#Calculating the derivatives of J wrt to w and b

def gradient(x,y,w,b):
    m = x.shape[0]
    
    dj_dw = 0
    dj_db = 0
    
    for i in range(m):
        f_wb = w * x[i] + b
        dj_dw += (f_wb - y[i]) * x[i]
        dj_db += f_wb - y[i]
    
    dj_dw /= m
    dj_db /= m
    
    return dj_dw, dj_db
```

```python
def grad_descent(x, y, w_in, b_in, alpha, iters):
   j_history = []
   p_history = []
    
   w = w_in
   b = b_in
   for i in range(iters):
       dj_dw, dj_db = gradient(x, y, w, b)
       cost = cost_func(x, y, w, b)
       j_history.append(cost)
       p_history.append([w,b])
       
       w -= alpha*dj_dw
       b -= alpha*dj_db
       
       if i%math.ceil(iters/10)==0:
           print(f'Iteration number: {i:4}\n'
                 f'\tCost: {cost:0.2e}'
                 f'\tw: {w:0.2e}'
                 f'\tb: {b:0.2e}'
                 f'\tdj_dw: {dj_dw:0.2e}'
                 f'\tdj_db: {dj_db:0.2e}\n')
    
   return w, b, j_history, p_history
        
```

```python
w_fin, b_fin, j_hist, p_hist = grad_descent(x_train, y_train,w, b, alpha, iters)
print(f'Final w and b are: {w:4.4f} and {b:4.4f}')
```

```python
def multi_cost(x, y, w, b):
    
    m = x.shape[0]
    cost = 0
    
    for i in range(m):
        cost += ((np.dot(x[i], w) + b) - y[i]) **2
        
    cost /= 2*m
    
    return cost
```

```python
cost = multi_cost(X_train, y_train, w_init, b_init)
print(cost)
```

```python
def multi_gradient(x, y, w, b):
    m, n = x.shape
    
    dj_dw = np.zeros(n)
    dj_db = 0
    
    for i in range(m):
        erro = (np.dot(x[i], w) + b) - y[i]
        for j in range(n):
            dj_dw[j] += erro * x[i, j]
        dj_db += erro
    
    dj_dw /= m
    dj_db /= m
    
    return dj_dw, dj_db
```

```python
def multi_gradient_descent(x, y, w_in, b_in, alpha, num_iters=10000):
    w = copy.deepcopy(w_in)
    b = b_in
    
    j_hist = []
    
    for i in range(num_iters):
        dj_dw, dj_db = multi_gradient(x, y, w, b)
        
        w -= alpha*dj_dw
        b -= alpha*dj_db
        
        cost = multi_cost(x, y, w, b)
        j_hist.append(cost)
        
        if i%math.ceil(num_iters/10) == 0:
            print(f'Iteration: {i}\n'
                  f'Cost: {cost:0.3f}\t'
                  f'w: {w}\t'
                  f'b: {b:0.3f}\n')
            
    return w, b, j_hist
```

```python
#FEATURE SCALING
def mean_normalization(x):
    X = copy.deepcopy(x)
    m,n = X.shape
    
    for i in range(n):
        col = X.iloc[:,i]
        meany = col.mean()
        mining = col.max() - col.min()
        
        for j in range(m):
            col.iloc[j] = (col.iloc[j] - meany)/ mining
            
    return X

def z_score_normalization(x):
    X = copy.deepcopy(x)
    m,n = X.shape
    
    for i in range(n):
        col = X.iloc[:,i]
        meany = col.mean()
        standy = col.std()
        
        for j in range(m):
            col.iloc[j] = (col.iloc[j] - meany)/standy
    
    return X

def mean_single(y):
    Y = copy.deepcopy(y)
    m = Y.shape[0]
    
    meany = Y.mean()
    mining = Y.max() - Y.min()
    
    for i in range(m):
        Y.iloc[i] = (Y.iloc[i] - meany)/mining
    
    return Y

def z_single(y):
    Y = copy.deepcopy(y)
    m = Y.shape[0]
    
    meany = Y.mean()
    standy = Y.std()
    
    for i in range(m):
        Y.iloc[i] = (Y.iloc[i] - meany)/standy
        
    return Y
```

```python
mean_X_train = mean_normalization(X_train)
z_scored_X_train = z_score_normalization(X_train)

mean_y_data = mean_single(y_train)
z_y_data = z_single(y_train)

print(type(y_train))
print(type(mean_y_data))
print(type(z_y_data))


print(type(X_train))
print(type(mean_X_train))
print(type(z_scored_X_train))
```
